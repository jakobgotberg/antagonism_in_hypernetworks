```python
import hypernetx as hnx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import warnings 
warnings.simplefilter('ignore')
```

```python
list_of_lists = [['book','candle','cat'],['book','coffee cup', 'man'],['coffee cup','radio', 'man']]

H = hnx.Hypergraph(list_of_lists)

plt.subplots(figsize=(4,4))
hnx.draw(H, with_edge_labels=False, with_node_labels=False, fill_edges=True)
```
