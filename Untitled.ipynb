```python
## Example from their github

NS = 'P' # P = Person
ES = 'M' # M = Meeting

HG = {
       ES+'1': {NS+'4': {'topics': ['A', 'B']}, 
                NS+'5': {'topics': ['A', 'C']}},
       
       ES+'2': {NS+'5':{'topics':['E', 'F']}, 
                NS+'6':{'topics':['F']}},
    
       ES+'3': {NS+'2':{'topics':['C', 'D']}, 
                NS+'3':{'topics':['B', 'C', 'D']}, 
                NS+'4':{'topics':['D']}},
    
       ES+'4': {NS+'1': {'topics':['C']}, 
                NS+'2': {'topics':['C']}, 
                NS+'3': {'topics':['C']}}
      }

edge_attributes = {
       ES+'1': {'interval': [0, 1]},
       ES+'2': {'interval': [2, 6]},
       ES+'3': {'interval': [3, 5]},
       ES+'4': {'interval': [7, 9]},
        }

node_attributes = {
       NS+'1': {'color': 'blue'},
       NS+'2': {'color': 'red'},
       NS+'3': {'color': 'green'},
        }


H = hnx.Hypergraph(HG, edge_properties = edge_attributes, node_properties = node_attributes)


H.add_incidence('M5', 'P3', weight = -1, topics = {'B', 'A'})

## where are the weights defined?
H.incidences.dataframe
H.adjacency_matrix().todense()
```

```python
# signed H?
import scipy.sparse as sp
def signed_adjacency_from_edge_sign(H, sign_prop="sign", use_incidence_weights=False):
    I, nodes, edges = H.incidence_matrix(index=True, weights=use_incidence_weights)

    # edge sign vector aligned with `edges` returned by incidence_matrix(index=True)
    s = np.array([H.get_properties(e, level=0, prop_name=sign_prop) for e in edges], dtype=float)
    S = sp.diags(s, format="csr")

    A = (I @ S @ I.T).tocsr()
    A.setdiag(0)
    A.eliminate_zeros()
    return A, nodes
    
edges = {"a": [1,2,3], "b": [2,3]}
edte_p = {"a":{"sign":-1}, "b":{"sign":1}} # This did nothing

H = hnx.Hypergraph(edges, edge_properties=edte_p)
#for v in H.edges['a']:  # nodes contained in hyperedge e
#        H.incidences[('a', v)].weight = -1
    
plt.subplots(figsize=(4,4))
hnx.draw(H, with_edge_labels=False, with_node_labels=False, fill_edges=True)

A, _ = signed_adjacency_from_edge_sign(H, use_incidence_weights=True)
A.todense()
#H.incidences.dataframe
```
